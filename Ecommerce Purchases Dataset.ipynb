# 18. How many people have a credit card that expires in 2020?

```python
ecomm_data['CC Exp Date']
```

```python
def fun():
    count = 0
    for date in ecomm_data['CC Exp Date']:
        if date.split('/')[1] == '20':
            count += 1
            
    print(count)
    
fun()       
```

```python
ecomm_data['Browser Info'].value_counts().head()
#ecomm_data['Browser Info'].value_counts().sort_values()
#the .value_counts() returns the count of unique elements of a column in descending order by default
# so if we use the method .sort_values(), it provides the result in ascending order.
```
